## Dataset Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('iris.csv', index_col = 0, header = 0)
df
```

```python
colors = {'Iris-setosa': 'tab:blue', 'Iris-versicolor': 'tab:red', 'Iris-virginica': 'tab:green'}
```

#### Histograms

```python
hist = df.hist(figsize = (10, 5), bins = len(df.index))
```

#### Scatter Plots

```python
fig = plt.figure(figsize = (12, 8))

for i, feature in enumerate(df.iloc[:, :-1]):
    ax = fig.add_subplot(221 + i)
    plt.scatter(df.index, df[feature], c = df[df.columns[-1]].map(colors))
    plt.title(feature)

plt.show()
```

```python
fig = plt.figure(figsize = (12, 8))

for i, feature in enumerate(df.iloc[:, :-1]):
    ax = fig.add_subplot(221 + i)
    plt.scatter(df.iloc[:, -1], df[feature], c = df[df.columns[-1]].map(colors))
    plt.title(feature)

plt.show()
```
